```python
import requests
from bs4 import BeautifulSoup
import json
import csv

def crawl_wikipedia_article(url):
    # 发送HTTP请求获取维基百科页面内容
    response = requests.get(url)

    # 检查请求是否成功
    if response.status_code == 200:
        # 使用BeautifulSoup解析HTML内容
        soup = BeautifulSoup(response.text, 'html.parser')

        # 根据维基百科页面的HTML结构找到文章内容的标签
        # 维基百科的文章内容通常在 <div id="mw-content-text"> 下
        article_content = soup.find('div', {'id': 'mw-content-text'})

        # 获取文章文本内容
        if article_content:
            article_text = article_content.get_text()
            
            print(article_content.get_text())

            # 将文章内容保存为JSON文件
            save_as_json(article_text)

            # 将文章内容保存为CSV文件
            save_as_csv(article_text)
        else:
            print("未找到维基百科文章内容的标签")
    else:
        print("请求失败，状态码：", response.status_code)

def save_as_json(article_text):
    # 将文章内容保存为JSON文件
    with open('戰爭.json', 'w', encoding='utf-8') as json_file:
        json.dump({'article_text': article_text}, json_file, ensure_ascii=False, indent=4)

def save_as_csv(article_text):
    # 将文章内容保存为CSV文件
    with open('戰爭.csv', 'w', encoding='utf-8', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['article_text'])
        csv_writer.writerow([article_text])

# 要爬取的维基百科页面的URL
wikipedia_url = "https://zh.wikipedia.org/zh-tw/%E6%88%98%E4%BA%89"

# 调用函数开始爬取文章并保存为JSON和CSV文件
crawl_wikipedia_article(wikipedia_url)
```
